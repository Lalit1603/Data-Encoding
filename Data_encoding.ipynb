```python
import pandas as pd
import numpy as np
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df
```

```python
df.info()
```

```python
cat_col = [col for col in df.columns if df[col].dtype=='category']
```

```python
cat_col
```

```python
new_df = df[cat_col]
```

```python
new_df
```

```python
for col in cat_col:
        print(f"{col} has {new_df[col].unique()} values. \n")
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(new_df[['sex']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
encoded_df
```

```python
final_df = pd.concat([new_df,encoded_df],axis = 1)
final_df
```

```python
final_df.drop(['sex_Female','sex'],axis = 1,inplace=True)
```

```python
final_df
```
